## Converting each rows in to tuple


```python
y = []
for i in range(len(df)):
    # Convert each value to a string before adding it to the tuple
    x = tuple(df.iloc[i].apply(str))
    y.append(x)

y
```

## Converting the tuple into text file

```python
file = open('Pizza_Sales.txt', 'w')
for tuple in y:
    file.write(str(tuple) + ',' + '\n')
file.close()
```
